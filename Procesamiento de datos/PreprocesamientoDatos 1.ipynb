# Binning


Binning es un proceso de transformación de variables numéricas continuas en 'contenedores' categóricos discretos, para análisis agrupados.

```python
plt.hist(df_num["age"])
plt.xlabel("age")
plt.ylabel("count")
plt.title("age bins")
plt.show()
```

Utilizamos numpy para crear tres categorias del mismo tamaño  <code>linspace(start_value, end_value, numbers_generated)</code>

```python
bins = np.linspace(min(df["age"]), max(df["age"]), 4)
bins
```

```python
categorias = ['Joven', 'EdadMedia', 'AdultosMayores']
```

Usamos cut para cuando se necesita segmentar y ordenar valores en contenedores

```python
pd.cut(df_num['age'], bins, labels=categorias, include_lowest=True )
```

```python
df_num['age-binned'] =  pd.cut(df_num['age'], bins, labels=categorias, include_lowest=True )
df_num.head(5)
```

```python
df_num["age-binned"].value_counts()
```

```python
plt.bar(categorias, df_num["age-binned"].value_counts())
plt.xlabel("age")
plt.ylabel("count")
plt.title("age bins")
plt.show()
```

# FEATURE SELECTION 

```python
y = np.asarray(df_cat["class"])
X = np.asarray(df_num4) #Solo estoy utilizando los atributos continuos
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Build a forest and compute the feature importances
forest = ExtraTreesClassifier(n_estimators=250,random_state=0)
forest.fit(X, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
