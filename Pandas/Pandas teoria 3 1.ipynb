# groupby

```python
df = pd.read_excel("Canada.xlsx")
df.head(2)
```

```python
df.groupby("Continente") #Me devuelve el objeto de la agrupación
```

```python
df.groupby("Continente").last()  #first Me devuelve un elemento de cada agrupación
```

```python
df[df["Tipo de region"]== "Developed regions"].head(1)
```

```python
df.groupby("Tipo de region").first() #first Me devuelve un elemento de cada agrupación
```

```python
df.groupby(["Continente","Tipo de region"]).first() #Podemos agrupar por más de una columna
```

```python
df.groupby(["Tipo de region","Continente"]).first() #Podemos agrupar por más de una columna
```

```python
df.groupby(["Continente"]).last() #Devuelve el último elemento de la agrupación
```

```python
df.groupby(["Continente"]).size() #Devuelve el tamaño de cada grupo de la agrupación
```

```python
df.groupby(["Tipo de region"]).size() #Devuelve el tamaño de cada grupo de la agrupación
```

```python
df.groupby(["Continente"]).count() #Cuenta el número de elementos que no son NaNs
```

```python
df.groupby("Continente")[1980].sum() #Agrupamos por continente y luego sumamos la migración de cada continente en el año 1980
```

```python
df.groupby("Continente")[[2008,2009,2010]].sum()
```

```python
df.groupby("Continente")[1980].max() # Para encontrar los valores máximos de cada continente en el año 1980
```

```python
df[df["Continente"]=="Europe"][["Pais",1980]]
```

```python
df.groupby("Continente")[[2010,2011,2012]].aggregate([min, max, sum])
```

```python
df.groupby("Continente")[list(range(1980,1988))].aggregate([min, max, sum ])
```

```python
df5 = df.groupby("Continente").agg({2010:['min','max','sum'],2011:['min','max'],2012:["sum"]})
df5
```
